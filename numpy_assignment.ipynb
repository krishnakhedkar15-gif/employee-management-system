## Assignment 1: Working with NumPy

Objective: Learn NumPy array creation and manipulation.

```
import numpy as np
```

### 1D Array Operations

```

arr_1d = np.arange(1, 21)
np.sum(arr_1d), np.mean(arr_1d), np.median(arr_1d), np.std(arr_1d), np.where(arr_1d > 10)
```

### 2D Array Operations

```

arr_2d = np.arange(1, 17).reshape(4, 4)
arr_2d, arr_2d.T, np.sum(arr_2d, axis=1), np.sum(arr_2d, axis=0)
```

### Random 3x3 Arrays

```

a = np.random.randint(1, 21, (3,3))
b = np.random.randint(1, 21, (3,3))
a + b, a - b, a * b, np.dot(a, b)
```

### Reshape and Slice

```

arr = np.arange(1, 13).reshape(3,4)
arr, arr[:2, 2:]
```
