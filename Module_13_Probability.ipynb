## 1. Basics of Probability

### a) Coin Toss Simulation
We simulate tossing a fair coin 10,000 times and compute experimental probabilities.

```

import random

tosses = 10000
heads = 0
tails = 0

for _ in range(tosses):
    if random.choice(['H', 'T']) == 'H':
        heads += 1
    else:
        tails += 1

print("Probability of Heads:", heads / tosses)
print("Probability of Tails:", tails / tosses)
```

### b) Rolling Two Dice
We estimate the probability of getting a sum of 7.

```

trials = 10000
count_sum_7 = 0

for _ in range(trials):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    if die1 + die2 == 7:
        count_sum_7 += 1

print("Probability of sum = 7:", count_sum_7 / trials)
```

## 2. Probability of At Least One 6 in 10 Rolls
We repeat experiments and count cases where at least one 6 appears.

```

def probability_at_least_one_six(trials=10000):
    success = 0
    for _ in range(trials):
        rolls = [random.randint(1, 6) for _ in range(10)]
        if 6 in rolls:
            success += 1
    return success / trials

print("Estimated Probability:", probability_at_least_one_six())
```

## 3. Conditional Probability & Bayes' Theorem
We simulate drawing balls with replacement.

```

colors = ['R']*5 + ['G']*7 + ['B']*8
previous_blue = 0
blue_then_red = 0

prev = random.choice(colors)

for _ in range(1000):
    curr = random.choice(colors)
    if prev == 'B':
        previous_blue += 1
        if curr == 'R':
            blue_then_red += 1
    prev = curr

print("P(Red | Previous Blue):", blue_then_red / previous_blue)
```

## 4. Discrete Random Variable
We generate a sample and compute statistics.

```

import numpy as np

values = [1, 2, 3]
probabilities = [0.25, 0.35, 0.4]

sample = np.random.choice(values, size=1000, p=probabilities)

print("Mean:", np.mean(sample))
print("Variance:", np.var(sample))
print("Standard Deviation:", np.std(sample))
```

## 5. Continuous Random Variable – Exponential Distribution

```

import matplotlib.pyplot as plt

data = np.random.exponential(scale=5, size=2000)

plt.figure()
plt.hist(data, bins=30, density=True)
plt.title("Histogram of Exponential Distribution")
plt.show()

plt.figure()
x = np.linspace(0, max(data), 1000)
pdf = (1/5) * np.exp(-x/5)
plt.plot(x, pdf)
plt.title("PDF of Exponential Distribution")
plt.show()
```

## 6. Central Limit Theorem Simulation

```

uniform_data = np.random.uniform(0, 1, 10000)

sample_means = []
for _ in range(1000):
    sample = np.random.choice(uniform_data, size=30)
    sample_means.append(np.mean(sample))

plt.figure()
plt.hist(uniform_data, bins=30)
plt.title("Uniform Distribution")
plt.show()

plt.figure()
plt.hist(sample_means, bins=30)
plt.title("Distribution of Sample Means")
plt.show()
```
